```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Initial_Sales_data=pd.read_csv("Untitled spreadsheet - Sheet1.csv")
```

```python
Initial_Sales_data
```

```python
x=Initial_Sales_data.drop("Quantity",axis=1)
```

```python
def convert_to_int(series):
    return series.str.replace(',', '').astype(float).round().astype(int)
x['Total Price'] = convert_to_int(x['Total Price'])
x['Unit Price'] = convert_to_int(x['Unit Price'])
```

```python
y=Initial_Sales_data["Quantity"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features=["Customer Name","Product Code","Sale Date","New_Date"]
```

```python
one_hot=OneHotEncoder(sparse_output=False)
transformer=ColumnTransformer([("one_hot",
                               one_hot,
                               categorical_features)],
                               remainder="passthrough")
transformed_x=transformer.fit_transform(x)
transformed_x
```

```python
pd.DataFrame(transformed_x)
```

```python
np.random.seed(42)
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.2)
```

```python
model.fit(x_test,y_test)
```
